```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
nba_df = pd.read_csv("nba.csv")
nba_df.head()
```

```python
nba_df.isnull().sum()
```

```python
nba_df["College"] = nba_df["College"].fillna(nba_df["College"].mode())
nba_df["Salary"] = nba_df["Salary"].fillna(nba_df["Salary"].mean())
```

```python
nba_df.dropna(inplace=True)
```

```python
nba_df.isnull().sum()
```

```python
def fun1(value):
    return int(value)
```

```python
nba_df['Age'] = nba_df['Age'].apply(fun1)
```

```python
nba_df.head()
```

```python
grouped_df = nba_df[["Number", "Age", "Weight", "Salary"]].groupby(nba_df["Height"])
```

```python
grouped_df.mean()
```

```python
median = grouped_df.median()
median
```

```python
min = grouped_df.min()
min
```

```python
max = grouped_df.max()
max
```

```python
std = grouped_df.std()
std
```

```python
iris = load_iris()
print(iris)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["species"] = iris.target_names[iris.target]
```
