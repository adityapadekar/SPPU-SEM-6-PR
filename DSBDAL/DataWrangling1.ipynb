````python
df['Fare'] = df['Fare'].ffill()
````

````python
df['Age'] = df['Age'].fillna(df['Age'].mean())
````

````python
df = df.drop(['Cabin'], axis = 1)
````

````python
df.isnull().sum()
````

````python
df["Age"].value_counts()
````

````python
def fun1(value):
    if (np.isnan(value) == False):
        return int(value)
````

````python
df["Age"] = df["Age"].apply(fun1)
````

````python
df.head()
````

````python
def fun2(value):
    if (value < 20):
        return "teenager"
    elif (value >= 20 and value < 40):
        return "young"
    elif (value >= 40 and value < 60):
        return "middle aged"
    else:
        return "senior citizen"
````

````python
df["Age"] = df["Age"].apply(fun2)
````

````python
def fun3(value):
    return 0 if value == 'male' else 1
````

````python
df['Sex'] = df['Sex'].apply(fun3)
````
